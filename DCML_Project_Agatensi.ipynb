## **Preparing training and test sets**

First, we remove the correct labels from the training and test sets and save these labels in a separate vector. Since features have very different ranges of values between them, we can apply the feature scaling to reduce these differences. We can use the min-max technique or the standard scaler. I prefer to use the standard scaler since the min-max restricts values between a specific range and then it is more sensitive to outliers. The standarsScaler is fitted only on the training set and the transformation is done one both training and test sets. This is done to avoid that the algorithm is influenced during training by the test set, distorting the model testing phase resulting in a model with higher performance metrics.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# First Config
X_train_first_config, X_test_first_config, y_train_first_config, y_test_first_config = train_test_split(X_first_config, y, test_size=0.3, random_state=42)
std_scaler_first_config = StandardScaler()
X_train_first_config_std = std_scaler_first_config.fit_transform(X_train_first_config)
X_test_first_config_std = std_scaler_first_config.transform(X_test_first_config)


# Second Config
X_train_second_config, X_test_second_config, y_train_second_config, y_test_second_config = train_test_split(X_second_config, y, test_size=0.3, random_state=42)
std_scaler_second_config = StandardScaler()
X_train_second_config_std = std_scaler_second_config.fit_transform(X_train_second_config)
X_test_second_config_std = std_scaler_second_config.transform(X_test_second_config)
```

Since I will use also a Recurrent Neural Network, I will use a different mechanism to shuffle data and to split the dataset between train and test set.  
I monitored the system by alterneting the normal state to the abnormal state (where a fault injection is performed).
Then, the dataset is organized as a chain of sequences where a sequence consists of data monitored while the system is in normal state followed by data monitored with the system under one specific attack.  
For RNNs, time sequences are of primary importance and then the order of datapoints inside the same sequence have to be kept unchanged.  
Then, I split the dataset into a list of sequences; subsequently I shuffle the list of sequences, and finally I will do the split between train and test sequences. After that I will apply the standard scaler on data.

```python
# Here there are some support functions that will be used to prepare the training and test sets for the RNN as explained before.

import random

def extract_sequences(input_dataset) -> list:
  """
  This function, given a dataset organized into sequences, where a sequence consists of data monitored while the system is in normal state (label 'injector' = 0) followed by data monitored with the system under attack (label 'injector' = 1),
  returns a list of sequences.
  :param input_dataset: dataset organized into sequences
  :return: list of sequences
  """
  sequences = []
  current_sequence = []
  column_names = input_dataset.columns.tolist()

  for i, row in input_dataset.iterrows():
      current_sequence.append(row)
      if row['injector'] == 1 and (i + 1 == len(input_dataset) or input_dataset.iloc[i + 1]['injector'] == 0):
          sequences.append(pd.DataFrame(current_sequence, columns=column_names))  # Save the sequence
          current_sequence = []

  # Add the last sequence if not empty
  if current_sequence:
      sequences.append(pd.DataFrame(current_sequence, columns=column_names))

  return sequences

def split_train_test(sequences: list, train_ratio: float=0.7, shuffle: bool=False , seed=None):
  """
  This function, given an input list, splits the list of sequences in two lists based on the train_ratio.
  :param sequences: list of sequences to split
  :param train_ratio: ratio of the train set
  :param shuffle: True is the sequences have to be shuffled between them
  :param seed: random seed to allow reproducibility
  :return: train_sequences, test_sequences
  """
  if seed is not None:
      random.seed(seed)

  if shuffle:  # shuffle the sequences
    random.shuffle(sequences)

  num_sequences = len(sequences)
  num_train_sequences = int(num_sequences * train_ratio)
  train_sequences = sequences[:num_train_sequences]
  test_sequences = sequences[num_train_sequences:]

  return train_sequences, test_sequences

def split_train_test_scale(input_dataset):
  """
  This function, given an input dataset, performs a split of the dataset into training and test sets, based on the concept of sequence. The returned training and test sets are also scaled.
  This function has to be used when you have a dataset organized into sequences and you want to keep unchanged the order of datapoints inside the same sequence.
  It is a useful function when you have to train and tast a Recurrent Neural Network where time series are involved.
  :param input_dataset: dataset to split
  :return: X_train_std, X_test_std, y_train, y_test, scaler
  """
  dataset_rnn = input_dataset.copy()
  dataset_rnn['injector'] = dataset_rnn['injector'].apply(lambda x: 0 if x == "None" else 1)

  train_sequences, test_sequences = split_train_test(extract_sequences(dataset_rnn), train_ratio=0.7, shuffle=True, seed=42)

  train_set_to_scale = pd.concat(train_sequences, ignore_index=True)
  test_set_to_scale = pd.concat(test_sequences, ignore_index=True)

  y_train = train_set_to_scale['injector']
  y_test = test_set_to_scale['injector']

  train_set_to_scale = train_set_to_scale.drop(columns=['injector'])
  test_set_to_scale = test_set_to_scale.drop(columns=['injector'])

  scaler = StandardScaler()
  X_train_std = scaler.fit_transform(train_set_to_scale)  # fit only on the train set and transform on both
  X_test_std = scaler.transform(test_set_to_scale)  # transform using the same scaler fitted on the train set

  return X_train_std, X_test_std, y_train, y_test, scaler
```

Now i use the three functions defined in the above module to do the split of the dataset into training and test sets for both configurations.

```python
# Scaling for the First Config
X_train_first_config_rnn_std, X_test_first_config_rnn_std, y_train_first_config_rnn, y_test_first_config_rnn, scaler_first_config_rnn = split_train_test_scale(dataset_first_config)

# Scaling for the Second Config
X_train_second_config_rnn_std, X_test_second_config_rnn_std, y_train_second_config_rnn, y_test_second_config_rnn, scaler_second_config_rnn = split_train_test_scale(dataset_second_config)
```

# **Training**

Since the goal of this project is to develop an anomaly detector that only detects known faults, I will only use supervised machine learning algorithms, as in general they are better than unsupervised ones in detecting these faults.
In particulat I will train:  
1) Random Forest: a bagging algorithm based on a forest of Decision Trees;  
2) eXtreme Gradient Boosting: a boosting algorithm whose base-learners are Decision Trees;  
3) Stacking of -> Random Forest + eXtreme Gradient Boosting + ADABoost;  
4) Voting of -> Random Forest + eXtreme Gradient Boosting + ADABoost;  
5) Recurrent Neural Network -> Gated Recurrent Unit (GRU)

## **5. Recurrent Neural Network -> Gated Recurrent Unit (GRU)**

GRU is a simplified version of LSTM (Long Short-Term Memory). This RNN has only a short-term states and then has less dense layers inside to control its gates (forget gate and input gate). In GRU, the output of the memory cell at each time step has the same value of its short-term state.

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import GRU, Dense, Dropout, Input, GlobalMaxPooling1D
from keras.regularizers import l1, l2
```

Reshape the input according to the constant variables that will be defined in each config.

```python
def create_sequences(input_dataset, labels, step_size, overlap):
    """
    Create subsequences (finestre temporali) that overlap, to give in input to the RNN.

    :param input_dataset: input dataset
    :param labels: labels corresponding to the input dataset
    :param step_size: lenght of each subsequence to create.
    :param overlap: number of overlapping datapoints between consegutive subsequences.

    :return: Tuple (X, y) where X is the array of subsequences, and y is the array of labels.
    """
    X = []
    y = []

    stride = step_size - overlap

    for i in range(0, len(input_dataset) - step_size + 1, stride):
        X.append(input_dataset[i:i + step_size])
        y.append(labels[i + step_size - 1])  # the label is the one of the final element in the subsequence

    X = np.array(X)
    y = np.array(y)

    return X, y
```

### **First Config**

```python
# CONSTANT VARIABLES
BATCH_SIZE_FC = 16
EPOCHS_FC = 100
TIME_STEPS_FC = 10
FEATURES_SHAPE_FC = X_train_first_config_rnn_std.shape[1]
NUM_OVERLAP_SAMPLES_FC = 4
```

```python
model_fc = Sequential([
    Input(shape=(TIME_STEPS_FC, FEATURES_SHAPE_FC)),
    GRU(units=64, activation='tanh', return_sequences=True, kernel_regularizer=l1(0.001)),
    Dropout(0.5),
    GRU(units=32, activation='tanh', return_sequences=True, kernel_regularizer=l2(0.001)),
    Dropout(0.5),
    GlobalMaxPooling1D(),
    Dense(units=32, activation='relu', kernel_regularizer=l1(0.001)),
    Dropout(0.5),
    Dense(units=16, activation='relu', kernel_regularizer=l2(0.001)), #8
    Dropout(0.5),
    Dense(units=1, activation='sigmoid')
])
```

Display the architecture of the Network:

```python
model_fc.summary()
```

```python
X_train_fc_rnn, y_train_fc_rnn = create_sequences(X_train_first_config_rnn_std, y_train_first_config_rnn, TIME_STEPS_FC, NUM_OVERLAP_SAMPLES_FC)

X_test_fc_rnn, y_test_fc_rnn = create_sequences(X_test_first_config_rnn_std, y_test_first_config_rnn, TIME_STEPS_FC, NUM_OVERLAP_SAMPLES_FC)
```

#### **Compile and train the model**

```python
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler

def exponential_decay_fn(epoch, lr):
    return lr * 0.1 ** (epoch / 20)

early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

initial_lr = 0.01

model_fc.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=initial_lr),
              loss='binary_crossentropy',
              metrics=['accuracy'])

lr_scheduler = LearningRateScheduler(lambda epoch: exponential_decay_fn(epoch, initial_lr))

rnn_clf_fc = model_fc.fit(
    X_train_fc_rnn,
    y_train_fc_rnn,
    batch_size=BATCH_SIZE_FC,
    epochs=EPOCHS_FC,
    validation_data=(X_test_fc_rnn, y_test_fc_rnn),
    callbacks=[early_stopping, lr_scheduler]
)
```

Shows the model accuracy graph and also the maximum accuracy achieved on the validation set. In fact, the model parameters are those for which the best accuracy was achieved on the validation set

```python
import matplotlib.pyplot as plt

plt.plot(rnn_clf_fc.history['accuracy'], label='Training Accuracy')
plt.plot(rnn_clf_fc.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
val_accuracy = max(rnn_clf_fc.history['val_accuracy'])
print(f"Best Validation Accuracy: {val_accuracy}")
```
